```python
import tqdm
import torch
import pygalmesh
import numpy as np

from model import DeepSDF
```

```python
model = DeepSDF()
model.load_state_dict(torch.load('chairtest.pth'))
```

```python
def chair(x):
    xyz = torch.from_numpy(np.array(x)).float()
    latent = torch.from_numpy(np.zeros(256)).float()
    with torch.no_grad():
        val = model(xyz, latent)
    return float(val.cpu().numpy()[0])
```

```python
ply = []
eps = 0.1
for x in tqdm.tqdm(np.arange(-1.0, 1.0, eps)):
    for y in np.arange(-1.0, 1.0, eps):
        for z in np.arange(-1.0, 1.0, eps):
            xyz = [x, y, z]
            if abs(chair(xyz)) < eps:
                ply.append(xyz)
```
